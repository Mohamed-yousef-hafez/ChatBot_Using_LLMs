```python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()
api_key=os.getenv('OPENAI_API_KEY')
client=OpenAI(api_key=api_key)

def ask_chatgpt(user_message):
  response=client.chat.completions.create(
      model="gpt-4o-mini",
      messages=[
          {"role":"system",
          "content":"you are a helpful assistant."}
          {"role":"user",
           "content":user_message}],tempreature=0.7

  )
  return response.choices[0].message.content
```

```python
!pip install gradio--quiet

import gradio as gr

def gradio_chat(user_message):
  return ask_chatgpt(user_message)

demo=gr.Interface(
      gf=gradio_chat,
      inputs=gr.Textbox(lines=2,placeholder="Ask me anything...",label="User Message"),
      outputs=gr.Textbox(label="ChatGPT Response"),
      title="Basic LLM ChatBot",
      description="A simple chatbot using openAI's API and gradio ut",
      allow_flagging="never"
  )
demo.launch()
```
